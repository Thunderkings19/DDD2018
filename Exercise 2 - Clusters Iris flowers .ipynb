```python
import seaborn as sns
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
df = pd.read_csv('iris.csv')
df.head()
```

```python
sns.scatterplot(x='PetalLengthCm', y='SepalLengthCm', style='Species', data=df)
```

There are three different species seen and the clusters that are seen with the naked eye is the Iris Setosa specy. All the three species are clustered though. The best cluster is the setosa, versicolor and virginica are a bit mixed up but also clustered.

```python
km = KMeans(n_clusters=3) #create a new k-means model with 3 clusters
X = df[['SepalLengthCm', 'PetalLengthCm']] #get the X variables from the dataframe
km = km.fit(X) #calculate the cluster centers
df['cluster'] = km.predict(X) #predict the clusters of each observation and store in the dataframe
df.head()
```

```python
sns.scatterplot(x='PetalLengthCm', y='SepalLengthCm', hue='cluster', data=df, s=50)
```

How does k-means perform: It does quite good, there are some outliers but for more then 80% it is correct!


```python
km = KMeans(n_clusters=5) #create a new k-means model with 3 clusters
X = df[['SepalLengthCm', 'PetalLengthCm']] #get the X variables from the dataframe
km = km.fit(X) #calculate the cluster centers
df['cluster'] = km.predict(X) #predict the clusters of each observation and store in the dataframe
df.head()
```

```python
sns.scatterplot(x='PetalLengthCm', y='SepalLengthCm', hue='cluster', data=df, s=50)
```
