```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('movies.csv')
df.head()
```

```python
df1 = pd.read_csv('ratings.csv')
df1.head()
```

```python
df_merge = pd.merge(df, df1, on='movieId') 
df_merge.head()
```

```python
df_pivot = pd.pivot_table(data=df_merge, index='userId', columns='title' ,values='rating')
df_pivot.head()
```

```python
sns.distplot(df_pivot['Star Wars: Episode I - The Phantom Menace (1999)'].dropna(), kde=False, bins=10)
plt.title('Ratings Star Wars: Episode I - The Phantom Menace (1999)')
plt.ylabel('Amount of chosen rating')
plt.xlabel('Rating (Stars)')
plt.show()
```
