```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('funda-amsterdam.csv')
df.head()
```

```python
df_subset = df[['price', 'bedrooms', 'rooms']]
df_subset.head()
```

```python
sns.pairplot(df_subset)
plt.show()
```

```python
sns.distplot(df['rooms'])
plt.show()
sns.distplot(df['price'])
plt.show()
sns.distplot(df['bedrooms'])
plt.show()
```

```python
df = df[df['price'] > 30000] #LET OP, MAAK EEN NIEUWE DATAFRAME EN GEBRUIK NIET JE HUIDIGE + MAAK ER EEN DATAFRAME VAN. 
df_subset = df[['price', 'bedrooms', 'rooms']]
```

```python
sns.lmplot(x='price', y='rooms', data=df_subset)
plt.xlabel('Price') 
plt.ylabel('Rooms')
plt.title('Price per amount of rooms on Funda Amsterdam')
plt.show()
```
